```python
a = np.array([1, 2, 3])
b = np.array([2, 4, 6])

plt.subplot(1, 2, 1)
# First[1,2] (1, 2, ..) indicates its ratio of graph
# last 1 = location of plot (1st)
plt.plot(a,b)

c = np.array([1, 2, 3])
d = np.array([5, 10, 15])

plt.subplot(1, 2, 2)
# First[1,2] (1, 2, ..) indicates its ratio of graph
# last 2 = location of plot (2nd)
plt.plot(c,d)

plt.show()
```

```python
x = np.array([10,20,30,50])
# starts from 0 degree angle
plt.pie(x)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
plt.pie(x,labels = labelss)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
plt.pie(x,labels = labelss,startangle = 90)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
explodee = [0.1,0.12,0.13,0.14]
plt.pie(x,labels = labelss,startangle = 90,explode =explodee)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
explodee = [0.5,0,0,0]
plt.pie(x,labels = labelss,startangle = 90,explode =explodee,shadow = True)
plt.show() 
```

```python
x = np.array([10,20,30,50])
labelss = ["MATHS", "HINDI", "SCIENCE", "ENGLISH"]
explodee = [0.5,0,0,0]
colorr = ['r','m','gold','g']
plt.pie(x,labels = labelss,startangle = 90,explode =explodee,shadow = True,colors = colorr)
plt.show() 
```
