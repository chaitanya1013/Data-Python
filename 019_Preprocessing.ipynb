```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler,Binarizer,Normalizer,OneHotEncoder,LabelEncoder
```

```python
np.random.seed(5) # here 5 is a specific seed value ( used 1, 100 or any other integer ) always get the same set of numbers
a = np.random.normal(0, 2, 10000) # np.random.normal(mean, standard_deviation, size)
b = np.random.normal(5, 3, 10000)
c = np.random.normal(-5, 5, 10000)

df = pd.DataFrame({
    'A' : a,
    'B' : b,
    'C' : c
})
df
```

```python
df.plot.kde()
```

```python
scaler = StandardScaler()
```

```python
scaled = scaler.fit_transform(df)
```

```python
df = pd.DataFrame(scaled, columns=['A','B','C'])
```

```python
df.plot.kde()
```

```python
minmax = MinMaxScaler()
mm_scaled = minmax.fit_transform(df)
```

```python
df = pd.DataFrame(mm_scaled,columns=['A','B','C'])
df.plot.kde()
```

```python
rscaler = RobustScaler()
```

```python
robust = rscaler.fit_transform(df)
df = pd.DataFrame(robust, columns=['A','B','C'])
```

```python
df.plot.kde()
```

```python
binary = Binarizer(threshold=5) # 5 means if (value > 5) convert to 1 ,(value <= 5) 0.
# binarizer is a tool that converts numerical data into binary data
a = np.array([[ 30, 10,  22], 
              [ 20,  2,  10],
              [ 33,  5, 5]])

binary1 = binary.transform(a)
binary1
```
